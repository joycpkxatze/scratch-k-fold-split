# preparation

import function that will be needed to manage the dataset and load the dataset

```python
import numpy as np
import pandas as pd
```

```python
#read csv
df = pd.read_csv("Iris.csv")                                    
df = df.rename(columns={"Species": "Label"})                    

df
```

# randomizing

because we want to do cross validation using k-fold and do the train and test process with the method we already choose, then we need to randomize the dataset **if the dataset is in ordered form** or just do the randomization part to make sure its randomized completely

```python
df = df.reindex(np.random.permutation(df.index))                                                          
print("Randomized data by randomizing the index : ")

df
```

as you can see, the dataset is completely randomized by now. then to make it easy to split the dataset, then we need to reset the index

```python
df = df.reset_index(drop=True)                                                                                   
print("Randomized data with reseted index: ")

df
```

now the index has been reset into ordered, and we have randomized dataset

# folds

now, we will make the folds. the code I'll write after this is spliting the dataset into folds manually, by counting and splitting one by one.

you can choose how many folds do you need to do validation (5 or 10 usually). if you're doing this manually like I did, then you need to count the index by yourself and make sure every folds **have the same size or ammount of data**

you can upgrade the code I write by myself with iteration or else

```python
#folds                                                              
fold1 = df.loc[0:29]                                            
fold2 = df.loc[30:59]
fold3 = df.loc[60:89]
fold4 = df.loc[90:119]
fold5 = df.loc[120:149]
```
